Funkcja do obliczenia czasów działanie operacji:

```r
df <- data.frame(size = numeric(0), vector = numeric(0), matrix = numeric(0))

for (x in seq(from=100,to=1000,by=100)){
    for (i in 1:10){
        vector1 <- runif(x,min = 0,max = 9)
        vector2 <- runif(x,min = 0,max = 9)
        matrix1 <- matrix(runif(x*x, min = 0, max = 9), nrow = x, ncol = x)
        start <- Sys.time()
        vector1*vector2
        end <- Sys.time()
        time1 <- end - start
        start <- Sys.time()
        matrix1 * vector1
        end <- Sys.time()
        time2 = end - start   
        df[nrow(df)+1,] <- c(x,time1,time2)
    }
}
```

Odczyt danych z pliku.

```r
results = read.csv("results.csv")
results
```

Pogrupowanie danych i obliczenie średnich wartości i odchyleń standardowych.

```r
library(dplyr)

res <- results %>%
  group_by(size) %>%
  summarise(mean_vector = mean(vector),
            mean_matrix = mean(matrix),
            sd_vector = sd(vector),
            sd_matrix = sd(matrix))
res
```

### Zad2

```r
library(ggplot2)
options(repr.plot.width=10, repr.plot.height=5)

ggplot(res, aes(x = size)) + 
  geom_point(aes(y = mean_vector), size = 4, color="darkblue") + 
  theme(text = element_text(size = 15), plot.title = element_text(hjust = 0.5)) +
  xlab("Rozmiar") +
  ylab("Czas [s]") +
  ggtitle("Średni czas mnożenia wektorów w zależności od rozmiaru")
```

```r
ggplot(res, aes(x = size)) +  
  geom_point(aes(y = mean_matrix), size = 4, color = "darkred") + 
  theme(text = element_text(size = 15), plot.title = element_text(hjust = 0.5)) +
  xlab("Rozmiar") +
  ylab("Czas [s]") +
  ggtitle("Średni czas mnożenia macierzy przez wektor\n w zależności od rozmiaru")
```

```r
ggplot(res, aes(x = size)) + 
  geom_point(aes(y = mean_vector, color = "Vector"), size = 4) + 
  geom_point(aes(y = mean_matrix, color = "Matrix"), size = 4) +
  theme(text = element_text(size = 15), plot.title = element_text(hjust = 0.5)) +
  xlab("Rozmiar") +
  ylab("Czas [s]") +
  ggtitle("Średni czas badanych funkcji") +
  scale_color_manual(values = c("darkred", "darkblue"), name = "Typ funkcji")
```
